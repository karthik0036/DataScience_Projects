```python
import requests

url = "https://vitapet.com/media/ga4l1d42/questions-before-puppy-900x600.jpg?anchor=center&mode=crop&width=1240&rnd=132503914680600000"
img_data = requests.get(url).content

with open("puppy.jpg", "wb") as f:
    f.write(img_data)

print("Image downloaded successfully!")
```

 # Resizing the image using Pillow Library

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("puppy.jpg")
img_resize = img.resize((200,200))

# plt.imshow(img)
# plt.axis("off")
```

```python
img_resize.save("puppy_resized.jpg")
```

```python
img_res = mpimg.imread("puppy_resized.jpg")
plt.imshow(img_res)
plt.show()
```
